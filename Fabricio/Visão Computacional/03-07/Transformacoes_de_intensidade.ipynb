```python
# importa os pacotes
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('2_7.png', cv2.IMREAD_GRAYSCALE)
altura, largura = img.shape
```

```python
n = np.arange(0,256)

# funcao para clarear a imagem
f_1 = (0.5294 * n + 120).astype(int)

plt.title('Funcao para clarear a imagem')
plt.plot(f_1)
plt.xlim([0,255])
plt.ylim([0,255])
plt.show()

img_1 = img.copy()
for i in range(0,altura):
  for j in range(0,largura):
    img_1[i,j] = f_1[ img_1[i,j] ]

hist = cv2.calcHist([img], [0], None, [256], [0,256])
hist_1 = cv2.calcHist([img_1], [0], None, [256], [0,256])

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.plot(hist)
plt.show()

plt.imshow(img_1, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.plot(hist_1)
plt.show()
```

```python
n = np.arange(0,256)

# funcao para escurecer a imagem
f_2 = (0.5882 * n).astype(int)

plt.title('Funcao para escurecer a imagem')
plt.plot(f_2)
plt.xlim([0,255])
plt.ylim([0,255])
plt.show()

img_2 = img.copy()
for i in range(0,altura):
  for j in range(0,largura):
    img_2[i,j] = f_2[ img_2[i,j] ]

hist = cv2.calcHist([img], [0], None, [256], [0,256])
hist_2 = cv2.calcHist([img_2], [0], None, [256], [0,256])

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.plot(hist)
plt.show()

plt.imshow(img_2, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.plot(hist_2)
plt.show()
```

```python
n = np.arange(0,256)

hist = cv2.calcHist([img], [0], None, [256], [0,256])
hist = hist / (altura*largura)
plt.plot(hist)
plt.show()

cdf = hist.cumsum()
cdf = cdf * 255
plt.plot(cdf)
plt.show()


plt.title('Funcao para equalizar o histograma')
plt.plot(cdf)
plt.xlim([0,255])
plt.ylim([0,255])
plt.show()

img_2 = img.copy()
for i in range(0,altura):
  for j in range(0,largura):
    img_2[i,j] = cdf[ img_2[i,j] ]

hist_2 = cv2.calcHist([img_2], [0], None, [256], [0,256])

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.plot(hist)
plt.show()

plt.imshow(img_2, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.plot(hist_2)
plt.show()
```
